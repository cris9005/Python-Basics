```python
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace('s', '')
loan_data['emp_length_int'] = loan_data['emp_length_int'].fillna(0)
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace("\+", '')
```

```python
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace("\+s", '')
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace('\<', '')
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace('+ years', '')
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace('< 1 year', str(0))
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace('n/a', str(0))
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace(' years', '')
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace(' year', '')
```

```python
#type(loan_data['emp_length_int'][0])
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' years', '')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' year', '')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' year', '')
```

```python
loan_data['emp_length_int'].unique()
```

```python
loan_data['emp_length_int'] = pd.to_numeric(loan_data['emp_length_int'])
```

```python
loan_data["earliest_cr_line_date"] = pd.to_datetime(loan_data["earliest_cr_line"], format = '%b-%y')
```

```python
type(loan_data['earliest_cr_line_date'][0])
```

```python
pd.to_datetime('2017-12-01') - loan_data['earliest_cr_line_date']
## calculates the diff in days for the loan dates to today...
```

```python
## Converts the time difference to months 
loan_data['mnths_since_earliest_cr_line'] = round(pd.to_numeric((pd.to_datetime('2017-12-01') - loan_data['earliest_cr_line_date'])/np.timedelta64(1, 'M')))
```

```python
loan_data['mnths_since_earliest_cr_line'].describe()
```

```python
## filtering data to find the months that are in zero, use loc to id columns with info and add on condition at the end
loan_data.loc[:, ["earliest_cr_line", 'earliest_cr_line_date', 'mnths_since_earliest_cr_line']][loan_data['mnths_since_earliest_cr_line'] < 0]

# Issue when the year was turned into 60s into 2060s. Origin starts in the 1960s which confused the conversion matrix
```

```python
loan_data.loc[:, ["earliest_cr_line", 'earliest_cr_line_date', 'mnths_since_earliest_cr_line']][loan_data['mnths_since_earliest_cr_line'] < 0]
```

```python
loan_data[ 'mnths_since_earliest_cr_line'][loan_data['mnths_since_earliest_cr_line'] < 0] = loan_data['mnths_since_earliest_cr_line'].max()
```
